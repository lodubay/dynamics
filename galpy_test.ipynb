```python
from galpy.potential import HernquistPotential, MiyamotoNagaiPotential, NFWPotential, plotRotcurve
import matplotlib.pyplot as plt
```

```python
bulge = HernquistPotential(a=0.1, normalize=0.2)
disk = MiyamotoNagaiPotential(a=0.2, b=0.05, normalize=0.4)
halo = NFWPotential(a=6, normalize=0.4)
```

```python
plotRotcurve(bulge+disk+halo, Rrange=[0.01,10], grid=1001, yrange=[0, 1.5])
bulge.plotRotcurve(Rrange=[0.01,10], grid=1001, overplot=True)
disk.plotRotcurve(Rrange=[0.01,10], grid=1001, overplot=True)
halo.plotRotcurve(Rrange=[0.01,10], grid=1001, overplot=True)
```
